## Understanding the Secondary Fan of a Polytope

```python
from refactored.gfan import *
```

```python
def create_polyhedron(A, b):
    """  Ax <= b  """
    inequalities = [[b[k]] + list(-A.rows()[k]) for k in range(A.nrows())]
    return Polyhedron(ieqs = inequalities)
```

```python
A = matrix([[-1, 0],
            [0, -1],
            [1,  1]])
b = free_module_element([0, 0, 1])
P = create_polyhedron(A, b)
```
